```python
# SISTEMA CON 3 CARACTERISTICAS
from sklearn import cluster
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

N_c = 6
color = ['y','m','k','r','b','c','g']

fig,axs = plt.subplots(3,4, figsize=(15,8))
X,t = make_blobs(n_samples = 100, n_features=3, centers = N_c, cluster_std=1, random_state=10)
#print(X)
axs[0][0].scatter(X[:,0],X[:,1], c=t,s=10)
axs[0][0].set_title('samples')
axs[1][0].scatter(X[:,1],X[:,2], c=t,s=10)
axs[1][0].set_title('samples')
axs[2][0].scatter(X[:,0],X[:,2], c=t,s=10)
axs[2][0].set_title('samples')
for max_iter,fig in zip([10,20,30],[1,2,3]):
  kmeans_clustering = KMeans(n_clusters=N_c, max_iter= max_iter, init='random')
  clusters = kmeans_clustering.fit_predict(X)
  print(clusters)
  axs[0][fig].set_title('iteraciones '+str(max_iter))
  axs[0][fig].scatter(X[:,0],X[:,1],c=clusters,s=10)
  axs[1][fig].set_title('iteraciones '+str(max_iter))
  axs[1][fig].scatter(X[:,1],X[:,2],c=clusters,s=10)
  axs[2][fig].set_title('iteraciones '+str(max_iter))
  axs[2][fig].scatter(X[:,0],X[:,2],c=clusters,s=10)
  clusters_center=kmeans_clustering.cluster_centers_
  axs[0][fig].scatter(clusters_center[:,0],clusters_center[:,1], marker='*',c='r',s=200)
  axs[1][fig].scatter(clusters_center[:,0],clusters_center[:,1], marker='*',c='r',s=200)
  axs[2][fig].scatter(clusters_center[:,0],clusters_center[:,1], marker='*',c='r',s=200)
```
